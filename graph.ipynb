```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

seqT = pd.read_csv('https://raw.githubusercontent.com/rnignon/Quic/main/seq.csv')
sendPending = pd.read_csv('https://raw.githubusercontent.com/rnignon/Quic/main/time.csv')
wnd = pd.read_csv('https://raw.githubusercontent.com/rnignon/Quic/main/wnd.csv')
df = pd.DataFrame(data = seqT.seqN.array, index = seqT.time, columns = ['seq'])
sendDf = pd.DataFrame(data = sendPending.test.array[:100], index = sendPending.time[:100], columns = ['test'])
```

```python
def plot_df(x, y, title="", xlabel='time', ylabel='seq', dpi=100):
    plt.figure(figsize=(27,9), dpi=dpi)
    plt.plot(x, y,  ".",color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(x=df.index, y=df.seq,
        title='Seq')
```

```python
def plot_df(x, y, title="", xlabel='time', ylabel='test', dpi=100):
    plt.figure(figsize=(27,9), dpi=dpi)
    plt.plot(x, y,  ".",color='tab:blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(x=sendDf.index, y=sendDf.test,
        title='Test')
```

```python
wndDf = pd.DataFrame(data = wnd.Wnd.array, index = wnd.Time, columns = ['Wnd'])

def plot_df(x, y, title="", xlabel='time', ylabel='wnd', dpi=100):
    plt.figure(figsize=(27,9), dpi=dpi)
    plt.plot(x, y,  ".",color='grey')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(x= wndDf.index, y=wndDf.Wnd,
        title='Wnd')
```
